# Radial distribution funcion

```python
rdf=[]
with open("md/tmp.rdf", "r") as f:
    for i in range(3):
        f.readline()
    for i in range(int(11)):
        f.readline()
        for j in range(50):
            rdf.append(f.readline().split())
```

```python
rdf=np.array(rdf, dtype="float").reshape(-1, 50, 4)
rdf_mean = rdf.mean(axis=0)
```

```python
plt.figure(figsize=(12, 9), facecolor='white')
plt.title("Radial Distribution Function", fontsize=12)
plt.plot(rdf_mean[:, 1], rdf_mean[:, 2], label="fix npt", linestyle="--", linewidth=2, color="blue")

plt.xlabel("$r$ / $\sigma$", fontsize=12)
plt.ylabel("$g_2(r)$", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.show()
```
